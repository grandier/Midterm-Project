### Deskripsi Data

Pada bagian ini, kami akan memberikan penjelasan terperinci tentang "Cirrhosis Patient Survival Prediction Dataset" yang tersedia di UCI Machine Learning Repository.

**Informasi Dataset:**
- **Nama Dataset:** Cirrhosis Patient Survival Prediction Dataset
- **Sumber Data:** UCI Machine Learning Repository
- **Tujuan:** Dataset ini dikumpulkan untuk menganalisis faktor-faktor yang berpengaruh terhadap kelangsungan hidup pasien dengan sirosis hati.
- **Jumlah Sampel:** Dataset ini terdiri dari 399 sampel pasien dengan sirosis hati.
- **Jumlah Fitur:** Terdapat 19 fitur yang berbeda untuk setiap sampel, yang mencakup informasi tentang karakteristik pasien, kondisi medis, dan parameter laboratorium.

**Fitur Utama:**
1. **Age:** Usia pasien saat diagnosis.
2. **Bilirubin:** Tingkat bilirubin dalam darah pasien.
3. **Cholesterol:** Kadar kolesterol dalam darah pasien.
4. **Albumin:** Tingkat albumin dalam darah pasien.
5. **Copper:** Kadar tembaga dalam serum pasien.
6. **Alkaline_Phosphotase:** Tingkat enzim fosfatase alkali dalam darah pasien.
7. **Platelets:** Jumlah trombosit dalam darah pasien.
8. **Prothrombin_Time:** Waktu pembekuan darah pasien.

**Variabel Target:**
- **Status:** Menunjukkan apakah pasien bertahan hidup (1) atau meninggal (2) dalam periode waktu tertentu setelah diagnosis.

**Tujuan Analisis:**
Tujuan dari analisis dataset ini adalah untuk memprediksi kelangsungan hidup pasien dengan sirosis hati berdasarkan fitur-fitur yang tersedia. Hasil analisis ini dapat memberikan wawasan yang berharga dalam pengembangan model prediktif untuk mendukung pengambilan keputusan klinis dalam manajemen pasien dengan sirosis hati.

**Referensi:**
- [Cirrhosis Patient Survival Prediction Dataset](https://archive.ics.uci.edu/dataset/878/cirrhosis+patient+survival+prediction+dataset-1) from UCI Machine Learning Repository.

Dengan menggunakan dataset ini, kami akan melanjutkan penelitian kami dengan melakukan analisis lebih lanjut untuk memahami faktor-faktor yang mempengaruhi kelangsungan hidup pasien dengan sirosis hati secara lebih mendalam.

```python
from ucimlrepo import fetch_ucirepo
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (8,4)
plt.rcParams['font.size'] = 8
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
```

```python
 # fetch dataset
cirrhosis_patient_survival_prediction = fetch_ucirepo(id=878)

# data (as pandas dataframes)
X = cirrhosis_patient_survival_prediction.data.features
y = cirrhosis_patient_survival_prediction.data.targets
```

```python
df = pd.concat([X, y], axis=1)
df.head()
```

Bisa dilihat, ada beberapa kolom yang tipe datanya masih salah, seperti **Cholesterol, Copper, Tryglicerides, Platelets**

```python
kolom = df.columns
kolom
```

## Data Pre-processing

```python
col_convert = ['Cholesterol', 'Copper', 'Tryglicerides', 'Platelets']
for i in col_convert:
    df[i].replace('NaNN', np.nan, inplace=True)
    mean_col = df[i].astype(float).mean()
    df[i] = df[i].astype(float)
    df[i].replace(np.nan, mean_col, inplace=True)
```
